```python
import requests as rq
```

```python
url = "https://www.indeed.com/jobs?q=data%20science&l=Remote&vjk=a2c008533625ab9d"
    
res = rq.get(url)
```

```python
res
```

```python
#Beautiful Soup is a Python package for parsing HTML and XML documents. It creates a parse tree for parsed pages that can be used to extract data from HTML, which is useful for web scraping

from bs4 import BeautifulSoup as bs # import the libraries

jobs = bs(res.content) # Getting content from the page
```

```python
jobs
```

```python
# getting elements by id
results = jobs.find(id="resultsCol")
```

```python
results
```

```python
#get elements by class
c_resuts = results.find_all("div", class_="messageContainer")
c_resuts
```

```python
# Using find_all()
c_resutsl = results.find_all("h2", class_="jobTitle")
job_titles = [t.text.strip() for t in c_resutsl]
```

```python
job_titles
```

```python
import pandas as  pd

df = pd.DataFrame(job_titles, columns=["Jon Title"])
```

```python
df
```

```python
df.head()
```

```python
# Now we save our data for further analysis. Thus we collect our data into a csv

df.to_csv("Indeed Remote Jobs First Page.csv")
```
